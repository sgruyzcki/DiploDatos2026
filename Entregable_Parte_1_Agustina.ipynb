Si analizados un lenguaje que solo usa 1 persona, ese sueldo representa a esa persona y no al lenguaje en si.

```
# 1. Para convertir los lenguajes separados por coma en una lista
def split_languages(languages_str):
  if not isinstance(languages_str, str):
    return []
  languages_str = languages_str.lower().replace('ninguno de los anteriores', '')
  return [lang.strip().replace(',', '') for lang in languages_str.split()]

# 2. Se aplica la función a la  tabla filtrada
df_filtrado.loc[:, 'cured_languages'] = df_filtrado['tools_programming_languages'].apply(split_languages)

# 3. Se separan los lenguajes en filas individuales
df_langs = df_filtrado.explode('cured_languages')

# 4. Se limpian filas vacías
df_langs = df_langs.dropna(subset=['cured_languages'])
df_langs = df_langs[df_langs['cured_languages'] != '']

# 5. Se obtiene las 10 de los más populares
top_10_names= df_langs['cured_languages'].value_counts().head(10).index

#6. Se incluye Work_seniority, salary_monthly y language
df_final = df_langs[df_langs['cured_languages'].isin(top_10_names)].copy()

# 7. Visualización
visual = df_langs['cured_languages'].value_counts().head(10).reset_index()
visual.columns= ['lenguaje', 'frecuencia']
print("Los 10 lenguajes más populares son:")
print(visual)
```

### Conteo de frecuencias de los lenguajes de programación

La columna que contiene información sobre los lenguajes de programación utilizados es `tools_programming_languages`. Sus valores son strings con los lenguajes seleccionados separados por comas.



```
df.tools_programming_languages[:3]
```

Las siguientes celdas de código separan estos lenguajes de programación y cuentan la frecuencia con la que aparecen.

No es necesario entender este código en profundidad, aunque sí es un buen ejercicio.

```
# Convert the comma-separated string of languages to a list of string.
# Remove 'ninguno de los anteriores' option, spaces and training commas.
def split_languages(languages_str):
  if not isinstance(languages_str, str):
    return []
  # Remove 'other' option
  languages_str = languages_str.lower()\
    .replace('ninguno de los anteriores', '')
  # Split string into list of items
  # Remove spaces and commas for each item
  return [lang.strip().replace(',', '')
          for lang in languages_str.split()]

# Create a new column with the list of languages
df.loc[:, 'cured_programming_languages'] = df.tools_programming_languages\
    .apply(split_languages)
if 'cured_programming_languages' not in relevant_columns:
    relevant_columns.append('cured_programming_languages')

# Duplicate each row of df for each programming language
# mentioned in the response.
# We only include in df_lang the columns we are going to analyze later, so we
# don't duplicate innecesary information.
df_lang = df.cured_programming_languages\
    .apply(pd.Series).stack()\
    .reset_index(level=-1, drop=True).to_frame()\
    .join(df[relevant_columns])\
    .rename(columns={0: 'programming_language'})
# Horrible programming style! But a lot of data science code can be written with
# as concatenations of functions (pipelines), and there's no elegant way of
# doing that on Python.
df_lang[:5]
```

En la columna `programming_language` se encuentra cada lenguaje por separado. Notar que si una respuesta contenía 3 lenguajes, como `"HTML, Javascript, Python"`, la fila ha sido replicada 3 veces. Por ello, hay tres filas con índice 1.

Para descubrir patrones entre varias variables, las herramientas más útiles son el Análisis de Correlación (Pearson) para variables numéricas y el análisis de Densidades condicionales para variables categóricas. Prácticamente, el uso del Gráfico de dispersión permite visualizar la naturaleza de estas relaciones, mientras que el filtrado por subpoblaciones específicas ayuda a aislar el efecto de variables confusoras permitiendo entender mejor que factores impulsan las diferencias salariales.

1. Herramientas Técnicas:
* Coeficiente de Pearson: para medir la fuerza y dirección de la relación lineal entre dos variables numéricas (Ej: Salario Bruto vs Neto). Es útil porque da un número entre -1 y 1, y cuanto más cercano a 1, más fuerte es la asociación positiva (si una sube la otra también).


*   Densidades Condicionales: permite comparar cómo cambia la distribución de una variable (Salario) según grupos de otras variables como nivel de estudio o género. Ayuda a determinar si las variables son independientes o si pertenecen a un grupo específico.



2. Herramientas prácticas:


*   Scatter Plot (Gráfico de Dispersión: es una herramienta para asociación, para ver si a medida que aumenta una variable, aumenta la otra.

 *   Mapas de color (Heatmpas): es útil para visualizar muchas correlaciones de forma simultánea. Permite identificar rápidamente qué pares de variables tienen asociaciones más fuertes mediante escala de colores.

3. Filtrados por Subpboblaciones:


*  Segementado: consiste en analizar grupos específicos, como filtrar solo pro gente con más de 5 años de experiencia. Para saber si los patrones cambian según el perfil del programador.





Para este ejercicio, el objetivo es centrarse en la relación entre dos variables numéricas. El objetivo es ver si el Salario Bruto y el Neto se mueven juntos y qué tan fuerte es ese vínculo.

1)  Selección de columnas y filas

```
# 1. Selección de columnas
columns = ['salary_monthly_BRUTO', 'salary_monthly_NETO']

# 2. Selección de filas: criterio de 300k a 10M para ambas variables
df_asociacion = df[
    (df['salary_monthly_NETO'] > 300000) & (df['salary_monthly_NETO'] < 10000000) &
    (df['salary_monthly_BRUTO'] > 300000) & (df['salary_monthly_BRUTO'] < 10000000)
].dropna(subset=columns)

print(f"Registros después de la limpieza: {len(df_asociacion)}")

# Para visualizar las primeras filas de la selección
df_asociacion[columns].head(10)
```

2. Herramientas Prácticas:


*  Gráfico de Dispersión: muestra la asociación y la distribución de cada variable.

```
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Diagrama de dispersión junto con histograma
g = sns.jointplot(data=df_asociacion,
                  x='salary_monthly_BRUTO',
                  y='salary_monthly_NETO',
                  kind='scatter',
                  alpha=0.3,
                  marginal_kws=dict(bins=25, fill=True))

# 2. Notación Científica
g.ax_joint.ticklabel_format(style='sci', axis='both', scilimits=(0,0))

# 3. Etiquetas y Título
g.set_axis_labels('Salario Mensual Bruto (ARS)', 'Salario Mensual Neto (ARS)', fontsize=12)

# Para el título en jointplot se usa plt.suptitle porque hay varios sub-gráficos
plt.suptitle('Asociación entre Salario Bruto y Neto', fontsize=15, y=1.05)

```

Explicación:
Para analizar la relación entre el salario bruto y neto, se utilizó un Joinplot. En el eje central se observa una asociación lineal positiva y fuerte, lo cual corresponde con un coeficiente de Pearson cercano a 1.  A medida que suben los valores, se ensancha en forma de embudo, lo que indica que a sueldos más altos, las retenciones impositivas.  

Los histogramas marginales revelan que ambas variables presentan un marcado sesgo positivo, concentrando a la mayor parte de la población en rangos salariales medios-bajos. Este abordaje confirma que el salario bruto es la mejor variable que explica el comportamiento del salario neto, y que la masa salarial se encuentra en el primer tercio del gráfico.



3. Herramientas Teóricas:


*   Coeficiente de Pearson: para cuantificar la asociación, calculamos el coeficiente que mide la fuerza de la relación lineal.


```
correlacion = df_asociacion['salary_monthly_BRUTO'].corr(df_asociacion['salary_monthly_NETO'])
print(f"Coeficiente de correlación de Pearson: {correlacion:.4f}")
```

El coeficiente de Peardon obtenido es 0,94, lo que indica una correlación lineal positiva fuerte. Esto significa que el Salario Bruto explica casi en su totalidad el comportamiento del Salario Neto. En términos práctico, existe una dependencia directa, a medida que aumenta la remuneración bruta, el impacto en el bolsillo (neto) es proporcionalmente ascendente.

No llega a ser 1, debido a que la relación entre ambos salarios no es estrictamente fija. Factores como cargas impositivas, introducen la varibabilidad.

**2. Herramientas Visuales y Modelos Utilizados**
Para estudiar el comportamiento de los datos se aplicaron las siguientes herramientas:

*  Gráficos de Dispersión: utilizados para descubrir patrones de asociación entre la experiencia acumulada y remuneración neta.
*   Estimación de densidad de Kernel (KDE): utiliza en los gráficos curvas para comparar como se distribuye el salario según el nivel de estudio. Esta herramienta permite visualizar la probabilidad de que un individio pertenezca a un rango salarial determinado.

```
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Variables
num_vars = ['salary_monthly_NETO', 'salary_monthly_BRUTO', 'profile_years_experience']
cat_vars = ['profile_studies_level', 'profile_gender']

# Etiqueta de los  ejes
mapa_nombres = {
    'salary_monthly_NETO': 'Salario Neto',
    'salary_monthly_BRUTO': 'Salario Bruto',
    'profile_years_experience': 'Años de Experiencia'
}

# Sueldos entre (300k a 10M) y quedamos con los 3 niveles de estudio más frecuentes
top_levels = df['profile_studies_level'].value_counts().head(3).index
df_plot = df[
    (df['salary_monthly_NETO'] > 300000) & (df['salary_monthly_NETO'] < 10000000) &
    (df['profile_studies_level'].isin(top_levels))
].copy()

# Renombramos las columnas en este DataFrame temporal para el gráfico
df_plot = df_plot.rename(columns=mapa_nombres)

# Visualización parte 1: densidades. Corresponde a la distribución de los datos
fig, axes = plt.subplots(1, 2, figsize=(16, 5))
plt.suptitle('Estudio de Distribución Salarial según Nivel de Estudio', fontsize=14, y=1.05)

colores = 'bright' # Tonos visibles

for i, col in enumerate(['Salario Neto', 'Salario Bruto']):
    sns.kdeplot(data=df_plot, x=col, hue='profile_studies_level',
                fill=True, common_norm=False, ax=axes[i], palette=colores)
    axes[i].ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    axes[i].set_title(f'Densidad de {col}')
    axes[i].set_xlabel('Monto (ARS)')

plt.tight_layout()
plt.show()

# Visualización parte 2: dispersión. Corresponde al comportamiento y patrones entre variables
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_plot,
                x='Años de Experiencia',
                y='Salario Neto',
                hue='profile_studies_level',
                alpha=0.4,
                palette=colores,
                s=40)

plt.title('Relación Experiencia vs Salario por Nivel de Estudio', fontsize=14)
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', title='Nivel de Estudio')
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```

**3. Comportamiento en los gráficos**
A partir de las visualizaciones obtenidas, se pueden extraer las siguientes conclusiones:

## b) Asociación

* Necesitamos decidir si sacar o no la columna de salario bruto. Para hacer la encuesta más simple.
¿Existe una correlación entre el salario bruto y el neto? Que abordaje y medidas usaría



Es recomendable eliminar una de las dos columnas, generalmente se conserva el Salario Neto, ya que refleja el poder adquisitivo real del encuestado para simplificar el formulario.

Existe una correlación lineal positiva casi perfecta entre ambas variables. Dado que el salario neto se calcula como un porcentaje del salario bruto (descontando aporte, obra social, ganancias, etc), el comportamineto de una explica casi en su totalidad la varianza de otra.

Para justificar, se emplea un abordaje mixto:
1. Abordaje visual: por medio del gráfico de dispersión con una línea de regresión. Permite visualizar rápidamente si los puntos forman una diagonal recta, lo que indicaría una relación lineal.

2. Medidas cuantitativas:

*   Covarianza: para confirmar la relación, si positiva, si sube el bruto, sube el neto.
*  Coeficiente de Correlación de Pearson (r): es la medida definitia para este caso, ya que cuantifica la relación lineal en una escala de -1 a 1. Un valor cercano a 1 indica que la columna extra es redundante.




```
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Límites de análisis
df_asociacion = df[
    (df['salary_monthly_NETO'] > 300000) & (df['salary_monthly_NETO'] < 10000000) &
    (df['salary_monthly_BRUTO'] > 300000) & (df['salary_monthly_BRUTO'] < 10000000)
].dropna(subset=['salary_monthly_NETO', 'salary_monthly_BRUTO'])

# 2. Cálculos
# Correlación de Pearson
correlacion = df_asociacion['salary_monthly_BRUTO'].corr(df_asociacion['salary_monthly_NETO'])
# Covarianza
covarianza = df_asociacion['salary_monthly_BRUTO'].cov(df_asociacion['salary_monthly_NETO'])

print("-" * 50)
print(f"MEDIDAS DE ASOCIACIÓN:")
print(f"Coeficiente de Correlación de Pearson: {correlacion:.4f}")
print(f"Covarianza: {covarianza:.2e}")
print("-" * 50)

# 3. Gráfico de Dispersión
plt.figure(figsize=(10, 6))
sns.regplot(data=df_asociacion,
            x='salary_monthly_BRUTO',
            y='salary_monthly_NETO',
            scatter_kws={'alpha':0.3, 'color': 'tab:blue'},
            line_kws={'color': 'red', 'linewidth': 2})

plt.title(f'Asociación Lineal: Bruto vs Neto (r = {correlacion:.4f})', fontsize=14)
plt.xlabel('Salario Mensual Bruto (ARS)')
plt.ylabel('Salario Mensual Neto (ARS)')
plt.ticklabel_format(style='sci', axis='both', scilimits=(0,0))
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

Se justifica eliminar el Salario Bruto.

Al visualizar la relación entre el Salario Bruto y Salario Neto mediante un gráfico de dispersión, se observa una franja de puntos que siguen la diagonal.

Esta respaldado numéricamente por el coeficiente de Correlación de Pearson r= 0.94, que es esperable, ya que el salario ento es una desviación matemática directa del bruto.



## c) Densidad condicional

Estudie la distribución del salario según el nivel de estudio.

Separe la población según el nivel de estudio (elija dos subpoblaciones numerosas) y grafique de manera comparativa ambos histogramas de la variable `'salary_monthly_NETO'`
¿Considera que ambas variables son independientes?
¿Qué analizaría al respecto?

Calcule medidas de centralización y dispersión para cada subpoblación






Se seleccionaron los perfiles con nivel de estudios Universitarios y Terciario, por ser dos de los grupos con mayor representatividad en la muestra.

```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Selección de nivel de estudios
subpoblaciones = ['Universitario', 'Terciario']

df_condicional = df[
    (df['salary_monthly_NETO'] > 300000) & (df['salary_monthly_NETO'] < 10000000) &
    (df['profile_studies_level'].isin(subpoblaciones))
].copy()

# 2. Histogramas comparativos
plt.figure(figsize=(12, 6))
sns.histplot(data=df_condicional,
             x='salary_monthly_NETO',
             hue='profile_studies_level',
             stat='density',       # Usamos densidad para poder comparar grupos de distinto tamaño
             common_norm=False,    # Cada histograma suma 1 por separado
             kde=True,             # Le suma la curvita suave por encima
             bins=30,
             alpha=0.5,
             palette='bright')

plt.title('Histograma Comparativo de Salario Neto: Universitario vs. Terciario', fontsize=14)
plt.xlabel('Salario Mensual Neto (ARS)')
plt.ylabel('Densidad')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.show()

# 3. Centralización y Dispersión
medidas = df_condicional.groupby('profile_studies_level')['salary_monthly_NETO'].agg(
    Media='mean',
    Mediana='median',
    Desvio_Estandar='std',
    Minimo='min',
    Maximo='max'
)

# Rango Intercuartílico
medidas['RIC'] = df_condicional.groupby('profile_studies_level')['salary_monthly_NETO'].apply(lambda x: x.quantile(0.75) - x.quantile(0.25))

pd.options.display.float_format = '{:,.0f}'.format
print("MEDIDAS DE CENTRALIZACIÓN Y DISPERSIÓN:")
display(medidas)
```

Análisis del histograma:



*   Centralización: al observar las medidas calculadas, tanto la media como la mediana son superiores en la población con estudios Universitarios. Sin embargo, en el histograma, se observa que las barras de mayor concentración (la moda) está fuertemenete solapadas en rangos salariales medios-bajos.
*   Dispersión: el nivel Universitario presenta un desvío estándar y un rango intercuartílico mayores. En el gráfico, se evidencia por medio de la cola derecha más larga.



**¿Considera que ambas variables son independientes?**

No, las variables no son estadísticamente independientes. Si el nivel de estudio y el salario fueran independientes, la probabilidad de ganar un determinado sueldo sería la misma sin importar qué título se tenga, además los histogramas serían iguales.


**¿Qué analizaría al respecto?**

Al analizar datos, la depedencia se manifiesta no en el sueldo inicial o medio, sino en la capacidad de alcanzar los valores extremos superiores. El nivel universitario alcanza una dispersión con outliers positivos mucho mayor que el terciario, confirmando que la variable categórica (nivel de estudio) si condiciona la distribución y la varianza numérica (salario neto).

**Selección de Variables:**

Para estudiar la densidad conjunta condicional se eligieron las variables numéricas Años de Experiencia y Salario Mensual Neto, discriminando en color por la variable categórica género. Esta sección permite evaluar si la trayectoria salarial a lo largo de los años es equitativa entre géneros.

```
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Establecemos límites, con límite de experiencia
df_ejercicio_d = df[
    (df['salary_monthly_NETO'] > 300000) & (df['salary_monthly_NETO'] < 10000000) &
    (df['profile_years_experience'] < 50)
].copy()

# 2. Géneros con más respuestas
top_generos = df_ejercicio_d['profile_gender'].value_counts().head(3).index
df_ejercicio_d = df_ejercicio_d[df_ejercicio_d['profile_gender'].isin(top_generos)]

# 3. Graficos
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df_ejercicio_d,
                x='profile_years_experience',
                y='salary_monthly_NETO',
                hue='profile_gender',
                alpha=0.4,           # Transparencia clave para ver dónde se superponen
                s=40,                # Tamaño de los puntos
                palette='bright')    # Paleta de colores vivos

# 4. Formato
plt.title('Densidad Conjunta Condicional: Salario vs Experiencia según Género', fontsize=14)
plt.xlabel('Años de Experiencia')
plt.ylabel('Salario Mensual Neto (ARS)')
plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
plt.legend(title='Género', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
```

En la franja entre 0 y 5 años de experiencia, hay un solapamiento entre los distintos géneros, indicando que las bandas salariales están estadarizadas y no presenta una brecha de género drástica en el punto de partida.

A medida que aumenta los años de experiencia y la dispersión salarial se amplia, predominan los sueldos máximos en perfiles masculinos. En cambio, las mujeres aun con 15 años de experiencia tienen un límite superior más bajo en su potencial salarial.

Con más de 20 años de experiencia, existe una escacez de perfiles femeninos, lo cual puede deberse a una incersión tardía y reciente de las mujeres en el mercado, o a un mayor indice de abandono en el sector.
