**¿Que es Normalizar?** --> EJEMPLO

Este ejemplo toma un conjunto de datos [10, 20, 30, 40, 50] y realiza los siguientes pasos:

1.Calcula la media de los datos utilizando np.mean(data), que devuelve el valor 30.

2.Calcula la desviación estándar de los datos utilizando np.std(data), que devuelve el valor 14.14.

3.Normaliza los datos restando la media y dividiendo por la desviación estándar: (data - mean) / std.



```python
import numpy as np
from sklearn.preprocessing import StandardScaler

import numpy as np

# Conjunto de datos de ejemplo
data = np.array([10, 20, 30, 40, 50])

# Cálculo de la media y desviación estándar
mean = np.mean(data)
std = np.std(data)

# Normalización de los datos
normalized_data = (data - mean) / std

print("Datos originales:", data)
print("Datos normalizados:", normalized_data)

scaler = StandardScaler()

normalizado = scaler.fit_transform(data.reshape(-1, 1))

print(normalizado)
```

**RobustScaler** es una técnica de preprocesamiento de datos utilizada en análisis de datos y aprendizaje automático para escalar características numéricas en un conjunto de datos. A diferencia de otros métodos de escalamiento, como el escalado estándar **(StandardScaler)**, el RobustScaler es más robusto a los valores atípicos o outliers en los datos.

```python
Price_scaled =MinMaxScaler().fit_transform(Price)
Price_scaled_R = RobustScaler().fit_transform(Price)
```
