## **A. Análisis sin discriminar por seniority**

Se realiza la limpieza del dataset + estimación de brecha salarial neta + IC95%

**1. Estimación puntual:**

Diferencia de medias salariales

```
alpha = 0.05  # Nivel de significación (1 - alpha = 95% confianza)

# Calcular medias por grupo y desviaciones estándar
x_h = np.mean(groupA)  # Media hombres
x_m = np.mean(groupB)  # Media mujeres
s_h = np.std(groupA, ddof=1)  # Desviación estándar hombres
s_m = np.std(groupB, ddof=1)  # Desviación estándar mujeres
n_h = len(groupA)  # Tamaño muestral hombres
n_m = len(groupB)  # Tamaño muestral mujeres

# Tamaño de la muestra
print(f"Tamaño groupA (varones): {len(groupA)}")
print(f"Tamaño groupB (mujeres): {len(groupB)}")

# Estimación puntual
diff_medias = x_h - x_m
print(f"\nEstimación puntual (diferencia de medias): {diff_medias:,.2f}")
```

```
# Intervalo de confianza usando t de Welch (no asume varianzas iguales)
# Error estándar de la diferencia
se_diff = np.sqrt((s_h**2 / n_h) + (s_m**2 / n_m))

# Grados de libertad (aproximación de Welch)
df = ((s_h**2 / n_h + s_m**2 / n_m)**2) / (((s_h**2 / n_h)**2 / (n_h - 1)) + ((s_m**2 / n_m)**2 / (n_m - 1)))

# Valor crítico para el intervalo de confianza
t_crit = stats.t.ppf(1 - alpha / 2, df)

# Margen de error
margen_error = t_crit * se_diff

# Intervalo de confianza
ic_inferior = diff_medias - margen_error
ic_superior = diff_medias + margen_error
print(f"Intervalo de confianza al {100*(1-alpha)}%: [{ic_inferior:,.2f}, {ic_superior:,.2f}]")
```

El IC al 95% es [464,572.68, 702,801.10]. Como el intervalo no contiene el 0, hay evidencia estadísticamente significativa de que existe una diferencia salarial entre varones y mujeres cis. En promedio, los varones ganan entre 464k y 702k pesos más que las mujeres.

**¿Cómo se relaciona este IC con el test de hipótesis para diferencia de medias?**

Si el IC no contiene el 0, entonces 0 no es un valor plausible para la diferencia real entre los grupos, lo que equivale a rechazar $H_0$ en el test de hipótesis con ese mismo nivel de significancia. Un IC al 95% que excluye el 0 corresponde a un p-valor< 0.05.

#**Resolución del ejercicio 2.1**

 Test de hipótesis para muestras de distintos tamaños y varianzas no homogéneas.

**Consideraciones:**

Los datos corresponden a los salarios observados en la encuesta (groupA y groupB), que representan una muestra de la población de trabajadores tech en Argentina.

El modelo es el supuesto teórico que adoptamos para hacer inferencia: bajo $H_0$, el estadístico t de Welch sigue una distribución t de Student con los grados de libertad de Satterthwaite. Este modelo nos permite calcular la probabilidad de observar una diferencia tan grande como la encontrada si en realidad no hubiera brecha (p-valor). Se tiene en cuenta que los datos nunca "prueban" nada por sí solos, es el modelo el que nos da el marco para interpretar qué tan compatibles son los datos con  $H_0$ .

**Hipótesis Nula y Alternativa**

La hipótesis nula es lo que asumimos verdadero hasta que nos damos cuenta de lo contrario.

$H_0=$ μ_A - μ_B = 0 → los salarios medios de varones y mujeres son iguales

$H_1=$ μ_A - μ_B ≠ 0→ los salarios medios de varones y mujeres son distintos


Es un test bilateral, ya que no asumimos a priori cuál grupo gana más.

**Estadístico (Pivote)**
Se utiliza el estadístico t de Welch:


$$t = \frac{\bar{X}_A - \bar{X}_B}{\sqrt{\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}}}$$

Se elige este test en lugar de Student porque los tamaños de muestra son muy distintos y los desvíos estándar no son homogéneos entre grupos.

**Distribución bajo $H_0$**

Bajo $H_0$ este estadístico sigue una distribución t de Student con grados de libertad calculados por la aproximación de Satterthwaite:
$$df = \frac{\left(\frac{s_A^2}{n_A} + \frac{s_B^2}{n_B}\right)^2}{\frac{(s_A^2/n_A)^2}{n_A-1} + \frac{(s_B^2/n_B)^2}{n_B-1}}$$




**Nivel de significancia**

Se fija $a$ = 0.05, es decir, se acepta un 5% de probabilidad de rechazar $H_0$ cuando en realidad es verdadera (error de tipo I).

#**Resolución ejercicio 2.2**



*   El estadístico t indica cuántos desvíos estándar está la diferencia observada respecto de lo esperado bajo $H_0$.
*   El p-valor es la probabilidad de observar una diferencia tan extrema como la encontrada, asumiendo que $H_0$ es verdadera.
*   Como p-valor < 0.05, se rechaza $H_0$: hay evidencia estadística de diferencia salarial entre grupos.




```
# Test de hipotesis considerando varianzas distintas
resultado = ttest_ind(groupA, groupB, equal_var=False)

print(f"Estadístico t: {resultado.statistic:.4f}")
print(f"Grados de libertad: {resultado.df:.4f}")
print(f"P-valor: {resultado.pvalue:.10f}")

if resultado.pvalue < alpha:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia estadística de diferencia salarial entre grupos.")
else:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) >= alpha ({alpha}) → No se rechaza H₀")
    print("  No hay evidencia suficiente de diferencia salarial entre grupos.")

# Verificación: el IC debe coincidir con el del Ejercicio 1
ci = resultado.confidence_interval(confidence_level=1-alpha)
print(f"\nIC al {(1-alpha)*100:.0f}% (verificación): [{ci.low:,.2f}, {ci.high:,.2f}]")
```

Como complemento al test de Welch, se aplica Mann-Whitney U, una prueba no paramétrica que compara las distribuciones completas sin asumir normalidad, útil dado que los salarios presentan asimetría incluso tras la limpieza. Si ambos tests coinciden en rechazar $H_0$, el resultado es más robusto.

```
# Test Mann-Whitney U
stat_mw, p_mw = mannwhitneyu(groupA, groupB, alternative='two-sided')
print(f"Mann-Whitney U: estadístico={stat_mw:.2f}, p-valor={p_mw:.6f}")

if p_mw < alpha:
    print(f"→ p-valor ({p_mw:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia de diferencia entre las distribuciones.")
else:
    print(f"→ p-valor ({p_mw:.6f}) >= alpha ({alpha}) → No se rechaza H₀")
    print("  No hay evidencia de diferencia entre las distribuciones.")
```

**Conclusión:**

Con un nivel de significancia del 5%, hay evidencia estadística de que los salarios medios netos de varones cis y mujeres cis son distintos. La diferencia estimada es de $583.686,89; con un IC al 95% de [464.572.68 ; 702.801,10], lo que indica que los varones ganan en promedio más que las mujeres en la muestra analizada.

El IC coincide con el calculado en el Ej 1, confirmando la equivalencia entre ambos enfoques.


#**Resolución ejercicio 2.3**

**¿Qué es la potencia? ¿Cómo se interpreta el poder estadístico de un test?**

La potencia es la probabilidad de detectar una diferencia real cuando realmente existe, es decir, la probabilidad de rechazar $H_0$ cuando $H_0$ es falsa.

Una potencia baja significa un riesgo alto de cometer un error de tipo II (no detectar una diferencia real que si existe). Convencionalmente se busca una potencia de al menos 0,8.
Este análisis debería realizarse antes de recolectar los datos, para determinar el tamaño muestral necesario y asegurarse de que el estudio tenga suficiente capacidad para detectar el efecto de interés.

```
# Effect size de Cohen (desvío estándar pooled)
nA = len(groupA)
nB = len(groupB)
pooled_std = np.sqrt(((nA - 1) * groupA.std()**2 + (nB - 1) * groupB.std()**2) / (nA + nB - 2))
effect_size = (groupA.mean() - groupB.mean()) / pooled_std
ratio = nB / nA

print(f"Effect size (d de Cohen, pooled): {effect_size:.4f}")
print(f"Interpretación: ", end="")
if effect_size < 0.2:
    print("efecto muy pequeño")
elif effect_size < 0.5:
    print("efecto pequeño-moderado")
elif effect_size < 0.8:
    print("efecto moderado")
else:
    print("efecto grande")
print(f"Ratio (nB/nA): {ratio:.4f}")
print()

# Tabla de tamaños necesarios
resultados = []
for power_objetivo in [0.8, 0.9, 0.95]:
    n_necesario = tt_ind_solve_power(
        effect_size=effect_size,
        nobs1=None,
        alpha=0.05,
        power=power_objetivo,
        ratio=ratio,
        alternative='two-sided'
    )
    resultados.append({
        'Potencia objetivo': power_objetivo,
        'n mínimo groupA': int(n_necesario),
        'n actual groupA': nA,
        'n mínimo groupB': int(n_necesario * ratio),
        'n actual groupB': nB,
        '¿Suficiente?': '✅' if nA >= n_necesario else '❌'
    })

df_power = pd.DataFrame(resultados)
display(df_power)
```

```
analysis = TTestIndPower()
actual_power = analysis.power(
    effect_size=effect_size,
    nobs1=nA,
    ratio=ratio,
    alpha=alpha
)
print(f"Potencia actual del test: {actual_power:.4f}")
```

**Interpretación del análisis de potencia:**

Se emplea D de Cohen, para determinar que tan grande es la diferencia entre ambas muestras.
D de Cohen = 0.3175 corresponde a un efecto pequeño-moderado según la escala de Cohen, lo que indica que la diferencia salarial entre varones y mujeres cis existe, pero no es muy grande en términos relativos al desvío estándar combinado de ambos grupos.

En los tres niveles de potencia evaluados (0.80, 0.90, 0.95) los tamaños actuales de muestra superan ampliamente el mínimo requerido. La potencia actual es prácticamente 1.0, lo que significa que si la diferencia existe en la población, el test la detectaría casi siempre.



**¿Es suficiente para ser representativo de la tendencia general?**

Estadísticamente si, pero la muestra proviene de una encuesta voluntaria de sysarmy, lo que introduce sesgos de selección. No es aleatoria ni representativa de toda la industria tecnológica argentina, por lo que los resultados deben interpretarse con cautela.

**¿Es suficiente para un juicio por discriminación?**

No. A pesar de alta potencia, para uso legal se requeriría una muestra aleatoria y representativa, además de controlar otras variables como rol, seniority y empresa, para poder establecer causalidad y no solo asociación estadística.

```
alpha = 0.05  # Nivel de significación (1 - alpha = 95% confianza)

# Calcular medias y desviaciones estándar
x_h = np.mean(groupA_s)  # Media hombres
x_m = np.mean(groupB_s)  # Media mujeres
s_h = np.std(groupA_s, ddof=1)  # Desviación estándar hombres
s_m = np.std(groupB_s, ddof=1)  # Desviación estándar mujeres
n_h = len(groupA_s)  # Tamaño muestral hombres
n_m = len(groupB_s)  # Tamaño muestral mujeres

# Tamaño de la muestra
print(f"Tamaño groupA (varones): {len(groupA_s)}")
print(f"Tamaño groupB (mujeres): {len(groupB_s)}")

# Estimación puntual
diff_medias = x_h - x_m
print(f"\nEstimación puntual (diferencia de medias): {diff_medias:,.2f}")

# Intervalo de confianza usando t de Welch (no asume varianzas iguales)
# Error estándar de la diferencia
se_diff = np.sqrt((s_h**2 / n_h) + (s_m**2 / n_m))

# Grados de libertad (aproximación de Welch)
df = ((s_h**2 / n_h + s_m**2 / n_m)**2) / (((s_h**2 / n_h)**2 / (n_h - 1)) + ((s_m**2 / n_m)**2 / (n_m - 1)))

# Valor crítico para el intervalo de confianza
t_crit = stats.t.ppf(1 - alpha / 2, df)

# Margen de error
margen_error = t_crit * se_diff

# Intervalo de confianza
ic_inferior = diff_medias - margen_error
ic_superior = diff_medias + margen_error
print(f"Intervalo de confianza al {100*(1-alpha)}%: [{ic_inferior:,.2f}, {ic_superior:,.2f}]")
```

La estimación puntual para el nivel Senior es de aproximadamente $492K de diferencia en favor de los varones. El IC al 95% no contiene el 0, por lo que hay evidencia estadísticamente significativa de brecha salarial también en este segmento.

```
# Test de hipotesis considerando varianzas distintas
resultado = ttest_ind(groupA_s, groupB_s, equal_var=False)

print(f"Estadístico t: {resultado.statistic:.4f}")
print(f"Grados de libertad: {resultado.df:.4f}")
print(f"P-valor: {resultado.pvalue:.10f}")

if resultado.pvalue < alpha:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia estadística de diferencia salarial entre grupos.")
else:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) >= alpha ({alpha}) → No se rechaza H₀")
    print("  No hay evidencia suficiente de diferencia salarial entre grupos.")

# Verificación: el IC debe coincidir con el del Ejercicio 1
ci = resultado.confidence_interval(confidence_level=1-alpha)
print(f"\nIC al {(1-alpha)*100:.0f}% (verificación): [{ci.low:,.2f}, {ci.high:,.2f}]")
```

```
# Test Mann-Whitney U
stat_mw, p_mw = mannwhitneyu(groupA_s, groupB_s, alternative='two-sided')
print(f"Mann-Whitney U: estadístico={stat_mw:.2f}, p-valor={p_mw:.6f}")

if p_mw < alpha:
    print(f"→ p-valor ({p_mw:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia de diferencia entre las distribuciones.")
```

Ambos tests (Welch y Mann-Whitney) coinciden en rechazar $H_0$ con p-valor< 0.05, lo que refuerza la robustez del resultado. La brecha salarial en el perfil Senior es estadísticamente significativa.

```
# Effect size de Cohen (desvío estándar pooled)
nA = len(groupA_s)
nB = len(groupB_s)
pooled_std = np.sqrt(((nA - 1) * groupA_s.std()**2 + (nB - 1) * groupB_s.std()**2) / (nA + nB - 2))
effect_size = (groupA_s.mean() - groupB_s.mean()) / pooled_std
ratio = nB / nA

print(f"Effect size (d de Cohen, pooled): {effect_size:.4f}")
print(f"Interpretación: ", end="")
if effect_size < 0.2:
    print("efecto muy pequeño")
elif effect_size < 0.5:
    print("efecto pequeño-moderado")
elif effect_size < 0.8:
    print("efecto moderado")
else:
    print("efecto grande")
print(f"Ratio (nB/nA): {ratio:.4f}")
print()

# Tabla de tamaños necesarios
resultados = []
for power_objetivo in [0.8, 0.9, 0.95]:
    n_necesario = tt_ind_solve_power(
        effect_size=effect_size,
        nobs1=None,
        alpha=0.05,
        power=power_objetivo,
        ratio=ratio,
        alternative='two-sided'
    )
    resultados.append({
        'Potencia objetivo': power_objetivo,
        'n mínimo groupA_s': int(n_necesario),
        'n actual groupA_s': nA,
        'n mínimo groupB_s': int(n_necesario * ratio),
        'n actual groupB_s': nB,
        '¿Suficiente?': '✅' if nA >= n_necesario else '❌'
    })

df_power = pd.DataFrame(resultados)
display(df_power)
```

La d de Cohen en Senior indica un efecto pequeño-moderado, lo que significa que la diferencia salarial existe pero no es muy pronunciada en términos relativos al desvío estándar combinado. La potencia actual es prácticamente 1.0, lo que refleja que con los tamaños muestrales disponibles el test si detectaría la diferencia casi siempre si existe en la población.

```
# Segmentacion por seniority: Senior
df_dev_s = df_developer[df_developer["work_seniority"] == "Senior"]
man_dev_s = df_dev_s.profile_g == 'Varón cis'
woman_dev_s = df_dev_s.profile_g == 'Mujer cis'

groupA_dev_s = df_dev_s[man_dev_s].salary_monthly_NETO
groupB_dev_s  = df_dev_s[woman_dev_s].salary_monthly_NETO

# Descripción de arreglos
print("=" * 55)
print("HOMBRES SENIOR DEVELOPER - SALARIO NETO")
print("=" * 55)
print(groupA_dev_s.describe().round(0))

print("\n" + "=" * 55)
print("MUJERES SENIOR DEVELOPER - SALARIO NETO")
print("=" * 55)
print(groupB_dev_s.describe().round(0))
```

```
alpha = 0.05  # Nivel de significación (1 - alpha = 95% confianza)

# Calcular medias y desviaciones estándar
x_h = np.mean(groupA_dev_s)  # Media hombres
x_m = np.mean(groupB_dev_s)  # Media mujeres
s_h = np.std(groupA_dev_s, ddof=1)  # Desviación estándar hombres
s_m = np.std(groupB_dev_s, ddof=1)  # Desviación estándar mujeres
n_h = len(groupA_dev_s)  # Tamaño muestral hombres
n_m = len(groupB_dev_s)  # Tamaño muestral mujeres

# Tamaño de la muestra
print(f"Tamaño groupA (varones): {len(groupA_dev_s)}")
print(f"Tamaño groupB (mujeres): {len(groupB_dev_s)}")

# Estimación puntual
diff_medias = x_h - x_m
print(f"\nEstimación puntual (diferencia de medias): {diff_medias:,.2f}")

# Intervalo de confianza usando t de Welch (no asume varianzas iguales)
# Error estándar de la diferencia
se_diff = np.sqrt((s_h**2 / n_h) + (s_m**2 / n_m))

# Grados de libertad (aproximación de Welch)
df = ((s_h**2 / n_h + s_m**2 / n_m)**2) / (((s_h**2 / n_h)**2 / (n_h - 1)) + ((s_m**2 / n_m)**2 / (n_m - 1)))

# Valor crítico para el intervalo de confianza
t_crit = stats.t.ppf(1 - alpha / 2, df)

# Margen de error
margen_error = t_crit * se_diff

# Intervalo de confianza
ic_inferior = diff_medias - margen_error
ic_superior = diff_medias + margen_error
print(f"Intervalo de confianza al {100*(1-alpha)}%: ({ic_inferior:.2f}, {ic_superior:.2f})")
```

```
# Test de hipotesis considerando varianzas distintas
resultado = ttest_ind(groupA_dev_s, groupB_dev_s, equal_var=False)

print(f"Estadístico t: {resultado.statistic:.4f}")
print(f"Grados de libertad: {resultado.df:.4f}")
print(f"P-valor: {resultado.pvalue:.10f}")

if resultado.pvalue < alpha:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia estadística de diferencia salarial entre grupos.")
else:
    print(f"\n→ p-valor ({resultado.pvalue:.6f}) >= alpha ({alpha}) → No se rechaza H₀")
    print("  No hay evidencia suficiente de diferencia salarial entre grupos.")

# Verificación: el IC debe coincidir con el del Ejercicio 1
ci = resultado.confidence_interval(confidence_level=1-alpha)
print(f"\nIC al {(1-alpha)*100:.0f}% (verificación): [{ci.low:,.2f}, {ci.high:,.2f}]")
```

```
# Test Mann-Whitney U
stat_mw, p_mw = mannwhitneyu(groupA_dev_s, groupB_dev_s, alternative='two-sided')
print(f"Mann-Whitney U: estadístico={stat_mw:.2f}, p-valor={p_mw:.6f}")

if p_mw < alpha:
    print(f"→ p-valor ({p_mw:.6f}) < alpha ({alpha}) → Se rechaza H₀")
    print("  Hay evidencia de diferencia entre las distribuciones.")
else:
    print(f"→ p-valor ({p_mw:.6f}) >= alpha ({alpha}) → No se rechaza H₀")
    print("  No hay evidencia de diferencia entre las distribuciones.")
```



Al restringir el análisis a perfiles Senior en empresas de tipo Developer, se controla parte de la heterogeneidad del conjunto de datos. Se observa que ambos test siguen rechazando $H_0$, lo que permite  reforzar la conclusión de que la brecha salarial entre varones y mujeres cis persiste incluso dentro de un segmento laboral más homogéneo. Esto hace menos plausible que la diferencia se deba solo a diferencia en el tipo de trabajo o industria.

```
# Effect size de Cohen (desvío estándar pooled)
nA = len(groupA_dev_s)
nB = len(groupB_dev_s)
pooled_std = np.sqrt(((nA - 1) * groupA_dev_s.std()**2 + (nB - 1) * groupB_dev_s.std()**2) / (nA + nB - 2))
effect_size = (groupA_dev_s.mean() - groupB_dev_s.mean()) / pooled_std
ratio = nB / nA

print(f"Effect size (d de Cohen, pooled): {effect_size:.4f}")
print(f"Interpretación: ", end="")
if effect_size < 0.2:
    print("efecto muy pequeño")
elif effect_size < 0.5:
    print("efecto pequeño-moderado")
elif effect_size < 0.8:
    print("efecto moderado")
else:
    print("efecto grande")
print(f"Ratio (nB/nA): {ratio:.4f}")
print()

# Tabla de tamaños necesarios
resultados = []
for power_objetivo in [0.8, 0.9, 0.95]:
    n_necesario = tt_ind_solve_power(
        effect_size=effect_size,
        nobs1=None,
        alpha=0.05,
        power=power_objetivo,
        ratio=ratio,
        alternative='two-sided'
    )
    resultados.append({
        'Potencia objetivo': power_objetivo,
        'n mínimo groupA': int(n_necesario),
        'n actual groupA': nA,
        'n mínimo groupB': int(n_necesario * ratio),
        'n actual groupB': nB,
        '¿Suficiente?': '✅' if nA >= n_necesario else '❌'
    })

df_power = pd.DataFrame(resultados)
display(df_power)
```

```
analysis = TTestIndPower()
actual_power = analysis.power(
    effect_size=effect_size,
    nobs1=nA,
    ratio=ratio,
    alpha=alpha
)
print(f"Potencia actual del test: {actual_power:.4f}")
```
