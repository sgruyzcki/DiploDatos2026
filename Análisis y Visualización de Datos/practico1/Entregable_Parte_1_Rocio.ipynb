## 1.Seleccionar columnas y filas relevantes

```
# complete here if you want to include more columns
relevant_columns = ['tools_programming_languages', 'salary_monthly_NETO','work_seniority','profile_years_experience','work_dedication','work_contract_type','work_province','sueldo_dolarizado']
```

```
# armo un nuevo df con las columnas que seleccioné
df2 = df[relevant_columns].copy()

# saco salarios menores o iguales a 0
df2 = df2.dropna(subset=['salary_monthly_NETO'])
df2 = df2[df2['salary_monthly_NETO'] > 0]

# elimino extremos con percentiles
p01 = df2['salary_monthly_NETO'].quantile(0.01)
p99 = df2['salary_monthly_NETO'].quantile(0.99)

df2 = df2[
    (df2['salary_monthly_NETO'] >= p01) &
    (df2['salary_monthly_NETO'] <= p99)
].copy()
```

```
df2.salary_monthly_NETO.describe()
```

```
# función para separar los lenguajes
def split_languages(languages_str):
    if not isinstance(languages_str, str):
        return []

    # paso a minúsculas y separo por coma
    languages = [lang.strip().lower() for lang in languages_str.split(",")]

    # elimino vacíos y "ninguno de los anteriores"
    languages = [lang for lang in languages if lang != "" and lang != "ninguno de los anteriores"]

    return languages

# creo una nueva columna con la lista de lenguajes
df2["cured_programming_languages"] = df2["tools_programming_languages"].apply(split_languages)

df2[["tools_programming_languages", "cured_programming_languages"]].head()
```

```
df_lang = df2.explode("cured_programming_languages").copy()

# renombro la columna para que quede más claro
df_lang = df_lang.rename(columns={"cured_programming_languages": "programming_language"})

# saco filas vacías
df_lang = df_lang.dropna(subset=["programming_language"])

df_lang.head()
```

Contar frecuencia de cada lenguaje

```
language_count = df_lang["programming_language"].value_counts().reset_index()
language_count.columns = ["language", "frequency"]

language_count.head(15)
```

## 2.Filtrar lenguajes relevantes

El siguiente código permite seleccionar sólo las filas de los 10 más frecuentes.

```
interesting_languages = language_count.head(10)["language"].tolist()
interesting_languages
```

```
filtered_df_lang = df_lang[df_lang["programming_language"].isin(interesting_languages)].copy()

filtered_df_lang.head()
```

## Respuesta a la pregunta

```
summary_lang = (
    filtered_df_lang
    .groupby("programming_language")["salary_monthly_NETO"]
    .agg(
        count="count",
        mean="mean",
        median="median",
        std="std",
        p25=lambda x: x.quantile(0.25),
        p75=lambda x: x.quantile(0.75),
        p90=lambda x: x.quantile(0.90)
    )
    .sort_values("median", ascending=False)
)

summary_lang
```

```
# ordenar de menor a mayor para que el más alto quede arriba al graficar horizontal
plot_df = summary_lang.sort_values("median", ascending=True)

plt.figure(figsize=(10,6))

# barras horizontales desde p25 hasta p75
plt.hlines(
    y=plot_df.index,
    xmin=plot_df["p25"],
    xmax=plot_df["p75"]
)

# punto en la mediana
plt.plot(
    plot_df["median"],
    plot_df.index,
    "o"
)

plt.xlabel("Salario neto mensual")
plt.ylabel("Lenguaje de programación")
plt.title("Mediana e intervalo intercuartílico del salario neto por lenguaje")
plt.grid(axis="x", alpha=0.3)
plt.show()
```

El punto muestra = la mediana del salario neto

La línea horizontal muestra = el intervalo intercuartílico, o sea desde p25 hasta p75. Se puede ver que tan dispersos estan los salarios netos

A tener en cuenta = Una misma persona puede haber declarado más de un lenguaje de programación, los grupos no son mutuamente excluyentes.

Para Java, la mediana del salario neto mensual es de $3,2 millones, lo que indica que el 50% de los casos se ubica por debajo de ese valor y el otro 50% por encima.

Bash/Shell, Python y TypeScript, tienen una mediana cercana a los $3 millones, por lo que conforman un segundo grupo de lenguajes asociados a salarios típicos relativamente altos
